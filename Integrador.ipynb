```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Se importan las librerias

```python
precio = 3
costo_fijo = 20000
mercado = (np.random.choice([1,2,3,4], 100, p=[0.3, 0.2, 0.4, 0.1])).tolist()
cv = (np.random.choice([0.15,0.2,0.22,0.3], 100, p=[0.1, 0.15, 0.20, 0.55])).tolist()


def map_demanda(etiqueta):
    demandas = {
        1:np.random.choice([5000, 6000], 1, p=[0.5, 0.5]),
        2:np.random.choice([7000, 8000], 1, p=[0.5, 0.5]),
        3:np.random.choice([9000, 10000], 1, p=[0.5, 0.5]),
        4:np.random.choice([11000, 12000], 1, p=[0.5, 0.5])
    }
    return demandas[etiqueta]
def ganancia(precio, demanda, costo):
    return precio * demanda - costo
def costo_variable(factor, demanda):
    return factor*demanda
    
  

keys = ['Escenario', 'Mercado','Demanda','Costo Fijo','Costo Variable','Ganancia']
```

Se definen las variables y las funciones basicas que retornan los valores requeridos
Se definen los 4 mercados y los costos variables y las demandas de cada mercado

```python
ss_dict = {
    'Escenario':[],
    'Mercado':[],
    'Demanda':[],
    'Costo Fijo':[],
    'Costo Variable':[],
    'Ganancia':[]
}
```

```python
spreadsheet = pd.DataFrame(ss_dict)
spreadsheet
```

```python
def step(spreadsheet, index, k):
    temp = []
    #escenario
    temp.append(index+1)
    #Mercado
    temp.append(mercado[index])
    #Demanda
    demanda = map_demanda(mercado[index])
    temp.append(demanda)
    #Costo Fijo
    temp.append(costo_fijo)
    #Costo Variable
    c_var = costo_variable(cv[index], demanda)
    temp.append(c_var)
    #Ganancia
    temp.append(ganancia(precio, demanda, c_var + costo_fijo ))
    df_temp = pd.DataFrame(dict(zip(keys, temp)), index=[0])
    return pd.concat([spreadsheet, df_temp])
```

Ahora se simulan 100 casos

```python
dspreadsheet = pd.DataFrame(ss_dict, dtype=int)
for i in np.arange(0,100):
    spreadsheet = step(spreadsheet, i, 0.5)

spreadsheet
```

Ahora se calcula la ganancia promedio y la desviacion estandar para luego realizar un analisis

```python
print('Ganancia Promedio: ' + str(spreadsheet['Ganancia'].mean()) + '\n' + 'Desviacion Estandar: ' + str(spreadsheet['Ganancia'].std()))
```

Ahora se calcula la Ganancia Promedio y desviacion estandar en cada escenario

```python
for mercado in [1,2,3,4]:
  print('Mercado ' + str(mercado) + ': Ganancia Promedio: ' + str(spreadsheet.loc[spreadsheet['Mercado'] == mercado]['Ganancia'].mean()) + ' Desviacion Estandar: ' + str(spreadsheet.loc[spreadsheet['Mercado'] == mercado]['Ganancia'].std()) + '\n')
```
